```python
### CODE USED TO FIND FRACTAL DIMENSTION
### initial python code provided by ChapGPT 4.0 and reviewed and used by author

import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def mm_normalization(p):
    return (p - p.min()) / (p.max() - p.min())

def z_score_normalization(data):
    return (data - np.mean(data)) / np.std(data)

def box_counting(data, box_sizes):
    N = len(data)
    #print("N>>",N)
    counts = []
    for box_size in box_sizes:
        num_boxes = int(np.ceil(N / box_size))
        non_empty_boxes = sum(
            1 for i in range(num_boxes) if np.any(data[i * box_size: (i + 1) * box_size]) 
        )
        counts.append(non_empty_boxes)
    inverted_box_sizes = [1/boxes for boxes in box_sizes]    # ties to 1/epsilon in the paper
    log_sizes = np.log(inverted_box_sizes)
    log_counts = np.log(counts)
    slope, intercept, r_value, p_value, std_err = stats.linregress(log_sizes, log_counts)
    return slope, r_value, p_value
    #print(f'{data_set_label}, {slope:.4f}, {r_value:.4f}, {p_value:.2e}')


# Function to compute the Hurst exponent using Rescaled Range (R/S) analysis
def compute_hurst_exponent(time_series, max_lag):
    lags = np.arange(2, max_lag)
    rs_values = []
    for lag in lags:
        segments = np.array_split(time_series, len(time_series) // lag)
        r_s_list = []
        for segment in segments:
            if len(segment) < lag:
                continue
            mean_adj = segment - np.mean(segment)
            cum_dev = np.cumsum(mean_adj)
            R = np.max(cum_dev) - np.min(cum_dev)
            S = np.std(segment) if np.std(segment) > 0 else np.nan
            if S > 0:
                r_s_list.append(R / S)
        rs_values.append(np.nanmean(r_s_list) if r_s_list else np.nan)
    # Remove NaN values
    valid_indices = ~np.isnan(rs_values)
    log_lags = np.log(lags[valid_indices])
    log_rs = np.log(np.array(rs_values)[valid_indices])
    # Fit a line to log-log plot using LinRegress
    # Typically people use polyplot, their difference is on the order of 10^-18, linregress was used for R-Value and P-Value
    Hslope, intercept, Hr_value, Hp_value, std_err = linregress(log_lags, log_rs)
    hurstLin.append(Hslope)
    return Hslope, Hr_value, Hp_value

def e2p(s):
    dex = s.index('e')
    return s[:dex]+"$x10^{"+s[dex+1:]+"}$"

### MAIN CODE

dATA= NSet.copy()
dATA_label=NLabels.copy()
setOfSetsLen=len(dATA)

slopeD,rvalD,D_weighted,slopeH,rvalH,H_weighted, siz = [],[],[],[],[],[],[]

scale=11                                        
boxSequence=[2**j for j in range(scale)]        

maxLag=100 

print("Creates data for the Hurst Exp. and Fractal Dimension Summary Table")
for t in range(setOfSetsLen):
    tLen=len(dATA[t])
    siz.append(tLen)
    ###
    _ = z_score_normalization(np.array(dATA[t]))
    Ds, Dr, Dp = box_counting(_, boxSequence)
    slopeD.append(Ds)
    rvalD.append(Dr)
    D_weighted.append(Ds*tLen)
    ###
    time_series = np.array(dATA[t])
    Hs, Hr, Hp = compute_hurst_exponent(time_series,maxLag)
    slopeH.append(Hs)
    rvalH.append(Hr)
    H_weighted.append(Hs*tLen)
    print(f'{dATA_label[t]} & {Hs:.4f} & {Hr:.4f} & {e2p(f'{Hp:.2e}')} & {Ds:.4f} & {Dr:.4f} & {e2p(f'{Dp:.2e}')} \\\\')


print("Sloep and R-Value calculations for weighted averages and stnd. dev., ranges, and median")
totLen=sum(siz)

D_weighted_average=sum(D_weighted)/totLen
H_weighted_average=sum(H_weighted)/totLen

varD=0
varH=0

for y in range(len(slp)):
    varD += ((slopeD[y] - D_weighted_average)**2)*siz[y]
    varH += ((slopeH[y] - H_weighted_average)**2)*siz[y]

varianceD = varD/totLen
varianceH = varH/totLen

print("D avg",D_weighted_average)
print("range D: ",min(slopeD),max(slopeD),"median",np.median(slopeD))
print("range D_Rvalue: ",min(rvalD),max(rvalD),"median",np.median(rvalD))
print("D variance and stnd dev.",varianceD,varianceD**.5)

print("H avg ",H_weighted_average)
print("range H: ",min(slopeH),max(slopeH),"median",np.median(slopeH))
print("range H_Rvalue: ",min(rvalH),max(rvalH),"median",np.median(rvalH))
print("H variance and stnd dev.",varianceH,varianceH**.5)

### Prelimenary test for Fractal Dimension for Box Sequence and Scale
print("Prelimenary test for Fractal Dimension for Box Sequence and Scale")
print("")
prelim = [0,3,32,35,44,47]
udoLin = [1, 2, 5, 10, 15, 20, 35, 55, 75, 105, 135, 190, 205]
print("Z Normalized - Upper half of the table")
for i in range(len(prelim)):
    _ = z_score_normalization(np.array(dATA[prelim[i]]))
    for k in range(9,14,2):
        s_G, r_G, pg = box_counting(_, [2**j for j in range(k)])
        s_E, r_E, pe  = box_counting(_, [math.floor(j**math.pi) for j in range(1,k+1)])
        s_PL, r_PL, ppl  = box_counting(_, udoLin[:k])
        print(f'{dATA_label[prelim[i]]+"_"+str(k)} & {s_G:.4f} & {r_G:.4f} & {s_E:.4f} & {r_E:.4f} & {s_PL:.4f} & {r_PL:.4f} \\\\')
print("")
print("MinMax Normalized - Lower half of the table")
for i in range(len(prelim)):
    _ = mm_normalization(np.array(dATA[prelim[i]]))
    for k in range(9,14,2):
        s_G, r_G, pg = box_counting(_, [2**j for j in range(k)])
        s_E, r_E, pe  = box_counting(_, [math.floor(j**math.pi) for j in range(1,k+1)])
        s_PL, r_PL, ppl  = box_counting(_, udoLin[:k])
        print(f'{dATA_label[prelim[i]]+"_"+str(k)} & {s_G:.4f} & {r_G:.4f} & {s_E:.4f} & {r_E:.4f} & {s_PL:.4f} & {r_PL:.4f} \\\\')
        

```

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy
import math
from scipy import stats
from scipy.stats import linregress


def zscore_normalize(data):
    return (data - np.mean(data)) / np.std(data)

def box_counting_dimension(data, box_sizes):
    """ Compute box-counting fractal dimension (log-log slope). """
    N = len(data)
    box_counts = []
    for box_size in box_sizes:
        num_boxes = int(np.ceil(N / box_size))
        non_empty_boxes = sum(
            1 for i in range(num_boxes) if np.any(data[i * box_size: (i + 1) * box_size])
        )
        box_counts.append(non_empty_boxes)
    log_box_sizes = np.log(1/box_sizes)
    log_box_counts = np.log(box_counts)
    slope, intercept, r_value, p_value, std_err = stats.linregress(log_box_sizes, log_box_counts)
    return slope, r_value, p_value, log_box_sizes, log_box_counts  # fractal dimension (slope) & correlation strength (r)

def plot_mini_loglog(NSetz, NLabelsz, a, b, cols):
    num_sets = b - a + 1
    rows = (num_sets + cols - 1) // cols  # Calculate rows to fit all plots
    fig, axes = plt.subplots(rows, cols, figsize=(cols * 4, rows * 4))
    axes = axes.flatten()
    box_sizes = np.array([2**i for i in range(11)])
    sloppy,areValues=[],[]
    for idx, i in enumerate(range(a, b + 1)):
        ax = axes[idx]
        data = np.array(NSetz[i])
        label = NLabelsz[i]
        # Z-score normalization
        norm_data = zscore_normalize(data)
        # Box-counting fractal dimension
        slope, r_value ,pp, lsizes, lcounts= box_counting_dimension(norm_data, box_sizes)
        # Plot log-log box-count graph
        N = len(norm_data)
        box_counts = []
        for box_size in box_sizes:
            num_boxes = int(np.ceil(N / box_size))
            non_empty_boxes = sum(
                1 for j in range(num_boxes) if np.any(norm_data[j * box_size: (j + 1) * box_size])
            )
            box_counts.append(non_empty_boxes)
        ax.plot(np.log(1/box_sizes), np.log(box_counts), 'o-', markersize=4)
        fnt=16
        ax.set_title(f'{label}\nD={slope:.4f}  r={r_value:.4f}', fontsize=fnt,fontweight='bold')
        ax.set_xlabel('log(1/ϵ)', fontsize=fnt-2)
        ax.set_ylabel('log(N(ϵ))', fontsize=fnt-2)
        ax.grid(visible=True, linestyle='--', alpha=0.9)
    for j in range(num_sets, len(axes)):
        fig.delaxes(axes[j])
    plt.tight_layout()
    plt.show()


### Plot first half, for table 1 or 2
print("D Log-Log Plots, Table 1 of 2")
a=0
b=(len(NSet)-1)//2
cols=4
plot_mini_loglog(NSet.copy(), NLabels.copy(), a, b, cols)

### Plot second half, for table 2 or 2
print("D Log-Log Plots, Table 1 of 2")
a=(len(NSet))//2
b=len(NSet)-1
cols=4
plot_mini_loglog(NSet.copy(), NLabels.copy(), a, b, cols)

```
